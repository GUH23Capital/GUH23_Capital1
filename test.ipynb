```python
import pandas as pd
from geopy.geocoders import Nominatim

# Load data from the original CSV file
df = pd.read_csv("transactions.csv")

# Initialize Nominatim API
geolocator = Nominatim(user_agent="geoapiExercises")

# Function to get location details for a given latitude and longitude
def get_location_details(lat, lon):
    location = geolocator.reverse(f"{lat},{lon}")
    address = location.raw['address']
    road = address.get('road', 'N/A')
    city = address.get('city', 'N/A')
    county = address.get('county', 'N/A')
    return road, city, county

# Add new columns for 'Road' and 'County' to the DataFrame
df['Road'], df['City'], df['County'] = zip(*df.apply(lambda row: get_location_details(row['Latitude'], row['Longitude']), axis=1))

# Save the updated DataFrame to a new CSV file with city information
df.to_csv("transactions_with_details.csv", index=False)

print("Data with city information saved to transactions_with_details.csv")
```

```python
import folium
import pandas as pd

# Load data from the CSV file
df = pd.read_csv("transactions_with_details.csv")

# Create a base map
m = folium.Map(location=[df['Latitude'].mean(), df['Longitude'].mean()], zoom_start=6)



# Add markers for each latitude and longitude
for index, row in df.iterrows():
    popup_text = f"<b>{row['Merchant Name']}</b><br>Timestamp: {row['Timestamp']}<br>Road:{row['Road']}<br>City: {row['City']}<br>County: {row['County']}"
    folium.Marker(
        location=[row['Latitude'], row['Longitude']],
        popup=folium.Popup(popup_text, max_width=200),
    ).add_to(m)

# Save the map to an HTML file (optional)
m.save("map.html")

# Display the map in a Jupyter Notebook (if using Jupyter)
m
```

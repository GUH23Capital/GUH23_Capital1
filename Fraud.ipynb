```python
import csv

with open('accounts.csv', 'r') as file:
    csv_reader = csv.DictReader(file)
    data = list(csv_reader)

    
# df_list = list(df.index)
# print(df_list)

fraud_url_list = []
for row in data:
    accountid = row['accountId']
    url = ""
    url += "https://sandbox.capitalone.co.uk/developer-services-platform-pr/api/data/fraud/transactions/accounts/"
    url+= accountid
    url += "/create"
    fraud_url_list.append(url)
    # print(url);
# print(url_list)

for x, row in enumerate(data):
    row['fraud_url'] = fraud_url_list[x]

with open('fraudlist.csv', 'w', newline='') as file:
    fieldnames = data[0].keys()
    csv_writer = csv.DictWriter(file, fieldnames=fieldnames)

    csv_writer.writeheader()
    csv_writer.writerows(data)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import csv

with open('frauds.csv', 'r') as file:
    csv_reader = csv.DictReader(file)
    data = list(csv_reader)


fraud_list = []
for row in data:
    fraudId = row['frauds_id']
    fraud_list.append(fraudId)

print(fraud_list)



print(fraudId)
fraudNumDic = {}
num = 0
for x in range(len(fraud_list)):
    if(fraud_list[x] not in fraudNumDic): fraudNumDic[fraud_list[x]] = 1
    else :fraudNumDic[fraud_list[x]] =fraudNumDic.get(fraud_list[x])+1 

   
times = list(fraudNumDic.keys())  
id = list(fraudNumDic.values())

plt.figure(figsize = (40, 10))
plt.bar(times,id)

plt.xlabel('accountId')
plt.ylabel('num')

plt.title("Number of transactions for each fraudulent account")
plt.show()
```
